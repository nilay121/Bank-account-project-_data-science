Convert the names of the states in state columns to lowercase and store it back in the same column

### Using if statement

```python
new_state=[]
ittr=len(df_account['state'])
for i in range(0,ittr):
    new_state.append(df_account['state'][i].lower())
```

```python
df_account['state']=new_state
df_account.head(2)
```

### Using lambda function

```python
df_account['state']=df_account['state'].apply(lambda x:x.lower())
```
